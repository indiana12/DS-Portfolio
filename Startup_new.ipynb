```python
import seaborn as sns #importing seaborn libraries
import matplotlib.pyplot as plt #importing matplot library
plt.scatter(x = 'City',y ='No. of Employees', data= startup_data, c='g', s=40)
plt.legend(['City','No. of Employees'], title='scatter_plot')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a histogram using seaborn
plt.figure()
sns.displot(startup_data['No. of Employees'])

# Display the plot
plt.show()
```

```python
plt.figure()
sns.countplot(x = 'No. of Employees',hue='Company', data=startup_data)
plt.show()
```
