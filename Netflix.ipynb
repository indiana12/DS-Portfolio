```python
# importing panda dataset in jupyter
# import seaborn library
import pandas as pd
import seaborn as sns

# read csv file
data = pd.read_csv(r"data/Netflix+Dataset.csv")

# display data
data

# display top 5
data.head()

# display bottom 5 records
data.tail()

# info() about the data
data.info()

# datatype of each col
data.dtypes

# removal of duplicate records
data.head()

data[data.duplicated()]

data.drop_duplicates(inplace=True)

# heck for the duplicates list in the functions
data[data.duplicated()]

# to check for the null values
data.isnull()

data.isnull().sum()

# seaborn heatmap
sns.heatmap(data.isnull())

# House of cards What is show id and director?
data.head()

data[data['Title'].isin(['House of Cards'])]

data[data['Title'].str.contains('House of Cards')]

# check data types
data.dtypes

# by release data and year
# data['Date_N'] = pd.to_datetime(data['Release_Date'])
data['Date_N'] = pd.to_datetime(data['Release_Date'], errors='coerce')

data.head()

data.dtypes

# year values
data['Date_N'].dt.year.value_counts()

# bar graph
data['Date_N'].dt.year.value_counts().plot(kind='bar')

# groupby
data.head(2)

data.groupby('Category').Category.count()

# countplot
sns.countplot(x='Category',
              data=data
              )
```
